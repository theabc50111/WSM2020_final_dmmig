```python
import pandas as pd
import numpy as np
```

### 73 columns 的one-hot encoding
/記得替換路徑/

```python
item_feature_73=pd.read_csv("../../recsys2022/item_features.csv")
item_feature_73["frequence"]=np.ones((len(item_feature_73),1))
item_feature_73
```

```python
item_feature_73 = item_feature_73.pivot_table(index="item_id", columns="feature_category_id",values="frequence")
item_feature_73 = (item_feature_73.fillna(0))
item_feature_73
```

### 讀取904 columns的one-hot encoding

```python
item_feature_904=pd.read_csv("../datasets/item_freq_vec.csv")
item_feature_904.set_index('item_id',inplace=True)
item_feature_904
```

### 將兩個dataframe合併為977 columns

```python
item_feature_977 = pd.merge(item_feature_904,item_feature_73, on='item_id')
item_feature_977
```
