````python
import pandas as pd
import numpy as np
````

## 讀檔
記得換掉讀檔路徑

````python
df = pd.read_csv("item_features.csv")
df['feature_category_id'] = df['feature_category_id'].apply(str)
df
````

___

## ↓對feature_category_id重新命名
```
e.g.
    id	fruit
0	1	Apple
1	1	Banana
2	1	Orange
3	1	Banana
```

會變成
```
    id	fruit
0	1	Apple_1
1	1	Banana_1
2	1	Orange_1
3	1	Banana_2
```


````python
s = df.groupby(['item_id', 'feature_category_id']).cumcount()
df['feature_category_id'] = (df.feature_category_id +'_'+ (s+1).astype(str) ).fillna(df.feature_category_id)
df
````


### 對groupby(['item_id','feature_category_id'])去數出現次數


````python
group = df.groupby(['item_id','feature_category_id']).count()
group.rename(columns={'feature_value_id': 'counts'}, inplace=True)
group
````

````python
group = group.pivot_table(index="item_id", columns="feature_category_id",values="counts")
group = group.fillna(0)
group.sort_index(ascending=True)

group
````

### 記得換掉存檔路徑

````python
group.to_csv('label_encoding.csv')
````
