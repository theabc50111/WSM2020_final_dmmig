```python
def inner_product(v1,v2):
    return sum(a*b for a, b in zip(v1, v2))
```

```python
def cosine_similarity(v1,v2):
    dot = sum(a*b for a, b in zip(v1, v2))
    norm_a = sum(a*a for a in v1) ** 0.5
    norm_b = sum(b*b for b in v2) ** 0.5
    return dot / (norm_a*norm_b)
```

```python
def eu_dist(v1,v2):
    return np.sqrt(sum((v1-v2)**2))
```

```python
def similarity_encoding(_session_df,_candi_arr,_item_df):
    '''
    given session,candidate(arr),item_df return similarity for every session to all candidate
    '''
    st = time.time()
    re_df = pd.DataFrame(columns=["session_id","item_id",'dot', 'cos', 'eu'])
    i = 0
    for idx in _session_df.index:
        for candi in _candi_arr:
            dot = inner_product(_session_df.loc[idx],_item_df.loc[candi])
            cos = cosine_similarity(_session_df.loc[idx],_item_df.loc[candi])
            eu = eu_dist(_session_df.loc[idx],_item_df.loc[candi])
            re_df.loc[i] = [(idx),(candi),dot,cos,eu]
            i+=1
    et = time.time()
    print("done, time:",et-st)
    return re_df
        
```
