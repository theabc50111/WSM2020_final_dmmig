```python
#normalize by col
cols = list(item_no_encode.columns[1:])
for col in cols:
    col_mean = item_no_encode[col].mean()
    col_std = item_no_encode[col].std()
    item_no_encode[col] = item_no_encode[col].apply(lambda x:(x-col_mean)/col_std)
item_no_encode.head()
```

### split month data for training and validation

```python
def sessions_time_split(_session_df,_datetime):
    '''
    input1: raw sessions df (must with date and session_id cols)
    input2: timestamp in *string* "####-##-##"
    return sessions_df[b4 timestamp],sessions_df[after timestamp]
    '''
    _session_df.date = pd.to_datetime(_session_df.date)
    time = pd.to_datetime(_datetime)
    df_b4 = _session_df[_session_df.date<time]
    df_after = _session_df[_session_df.date>=time]
    return df_b4,df_after
```

```python
train_sessions=pd.read_csv("../../recsys2022/train_sessions.csv")
train_sessions
```

```python
train_x,test_x = sessions_time_split(train_sessions,"2021-05-01")
train_x
```

```python
train_x.to_csv("../datasets/train_sessions_x.csv")
test_x.to_csv("../datasets/test_sessions_x.csv")
```
