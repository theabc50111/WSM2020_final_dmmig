```python
import pandas as pd
```

## read no encode csv

```python
item_no_encode = pd.read_csv('../datasets/no_encoding.csv')
#item_id = item_no_encode["item_id"]
#item_no_encode=item_no_encode.drop(columns=["item_id"])
item_no_encode.head()
```

```python
#normalize by col
cols = list(item_no_encode.columns[1:])
for col in cols:
    col_mean = item_no_encode[col].mean()
    col_std = item_no_encode[col].std()
    item_no_encode[col] = item_no_encode[col].apply(lambda x:(x-col_mean)/col_std)
item_no_encode.head()
```

## read onehot

```python
item_freq_encode = pd.read_csv('../datasets/item_freq_vec.csv')
item_freq_encode.head()
```

## make one hot + no encoding csv

```python
item_no_n_onehot = pd.merge(item_freq_encode,item_no_encode, on='item_id')
item_no_n_onehot.head()
```

```python
item_no_n_onehot.to_csv("../datasets/item_no_n_onehot.csv")
```
